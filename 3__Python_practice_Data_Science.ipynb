#### Calculator

```python
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))
op = input('Enter operator +, -, *, /: ')
if op == '+':
  print(num1 + num2)
elif op == '-':
  print(num1 - num2)
elif op == '*':
  print(num1 * num2)
elif op == '/':
  if num2 == 0:
    print('Not defined')
  else:
    print(num1 / num2)
```

```python
flag = True
while flag:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    op = input('Enter operator +, -, *, /: ')
    if op == '+':
      print(num1 + num2)
    elif op == '-':
      print(num1 - num2)
    elif op == '*':
      print(num1 * num2)
    elif op == '/':
        if num2 == 0:
          print('NOt defined!')
        else:
          print(num1 / num2)
    else:
      print('Invalid input operator! Try valid operator')
    con = input('Press 0 to Switch off the calculator or Press any key to Continue: ')
    if con == '0':
        flag = False
    else:
        flag = True
```

### Q. Create a calculator which works infinitely but stops as soon as user enter  as operator.

```python
flag = True
while flag:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    op = input('Enter operator +, -, *, /: ')
    if op == '0':
        flag = False
    else:
      if op == '+':
          print(num1 + num2)
      elif op == '-':
          print(num1 - num2)
      elif op == '*':
          print(num1 * num2)
      elif op == '/':
          if num2 == 0:
            print('Not defined')
          else:
            print(num1 / num2)
```

### WAP to check if a number is prime or not
- using for loop
- using while loop

```python
num = int(input('Enter a number:' ))
prime = True
for i in range(2, num):
    if num % i == 0:
        prime = False
if prime:
    print('Prime')
else:
    print('Not Prime')
```

#### Jump Statements
- break: as reaigning from the job permanently i.e. once you leave you won't come back.
  - As soon as  a loop gets a break statement, loop will be terminated immediately.
- Continue: as taking a half-day leave i.e.. If you leave for half day, you come back next day
  - As soon as a loop gets a continue statement, the current iteration stops and then the control moves on to next iteration of the looop
- return (will be coverd in functions)

```python
for i in range(1, 11):
  print(i)
```

```python
for i in range(1, 11):
  print(i)

  if i == 5:
    break
print('i outside the loop', i)

# # Dry Run
# i   p   ifCon   break
# 1   1    F
# 2   2    F
# 3   3    F
# 4   4    F
# 5   5    T      break
```
